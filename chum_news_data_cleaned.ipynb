```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('hdfc_news_data_chum.csv')
```

```python
data.head()
```

```python
data.drop(['web-scraper-order', 'web-scraper-start-url', 'pages', 'pages-href', 'Year', 'Year-href'], axis=1, inplace=True)
```

```python
data.head()
```

###  Spliting the 'Headlines' column

```python
data[['Time', 'Date', 'Source']] = data['Headlines'].str.split('|', expand=True)
```

### Cleaning up the columns

```python
data['Time'] = data['Time'].str.strip()
data['Date'] = data['Date'].str.strip()
data['Source'] = data['Source'].str.replace('Source:', '').str.strip()
```

### Remove the extracted part from the original 'Headliens' column

```python
data['Headlines'] = data['Headlines'].str.split('|').str[0].str.strip()
```

```python
data
```

```python
data.to_csv('chum_cleaned_data.csv', index=False)
```
