```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
```

## Loading the url

```python
url = 'https://in.investing.com/equities/hdfc-bank-ltd-historical-data?end_date=1706156264&st_date=1390674600'
```

## Using a headless browser to simulate a real browser

```python
options = webdriver.ChromeOptions()
options.add_argument('--headless') #Run chrome without openning a browser window
driver = webdriver.Chrome(options=options)
```

## Load the webpage and wait for some time to let the JavaScript load the content

```python
driver.get(url)
driver.implicitly_wait(10) # Wait for 10 seconds (custom adjustable)
```

## Get the page source after the JavaScript has loaded the content

```python
page_source = driver.page_source
```

## Close browser

```python
driver.quit()
```

## Parse the page source with BeautifulSoup

```python
soup = BeautifulSoup(page_source, 'html.parser')
```

## Find the table

```python
table = soup.find("table", class_="common-table medium js-table")

print(table)
```

## Convert the table to a dataframe

```python
df = pd.read_html(str(table))[0]
```

## Export the DataFrame to a CSV file

```python
df.to_csv('ten_years_stock_hdfc.csv', index=False)
```
